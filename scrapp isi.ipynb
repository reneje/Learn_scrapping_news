```python
def bersihsub(expr):

    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("  ","\n",hsl)
    return hsl
```

```python
judul = bersihsub(judul)
time = bersihsub(time)
content = bersihsub(content)
print(judul)
print(time)
print(content)
```

```python
import re

date = re.sub("K.*? - ","",time)
date = re.sub(",","",date)
date
```

```python
file = open("data_link.txt").read()
files = file.split("\n")
files
```

```python
links=[]
titles =[]
dates=[]
contents=[]
for link in files:
    print(link)
    page = requests.get(link)
    soup = BeautifulSoup(page.text,'html.parser')
    judul = soup.find(class_ = 'read__title')
    time = soup.find(class_ = 'read__time')
    content = soup.find(class_ = 'read__content')
    judul = bersihsub(judul)
    time = bersihsub(time)
    content = bersihsub(content)
    date = re.sub("K.*? - ","",time)
    date = re.sub(",","",date)
    links.append(link)
    titles.append(judul)
    dates.append(date)
    contents.append(content)
    print("Sukses")
```

```python
import pandas as pd
raw_data = {"links":links,"titles":titles,"dates":dates,"contents":contents}
data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
```
