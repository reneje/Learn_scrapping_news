```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/Users/rinijannati/anaconda/bin/chromedriver", chrome_options=options)

driver = webdriver.Chrome()

url = "https://katadata.co.id/search/cse/p2p%20lending/-/-/-/-/-/-/"
links=[]
driver.get(url)
hrefs = driver.find_elements_by_xpath("//*[@class='search-results']//*[@href]")
print('Number of results', len(hrefs))
for a in hrefs:
    link = a.get_attribute('href')
    print(link)
    links.append(link)
    print("sukses")

for i in range(2,10):
    url = "https://katadata.co.id/search/cse/p2p%20lending/-/-/-/-/-/-/"+str(i*10)
    driver.get(url)
    hrefs = driver.find_elements_by_xpath("//*[@class='search-results']//*[@href]")
    print('Number of results', len(hrefs))
    for a in hrefs:
        link = a.get_attribute('href')
        print(link)
        links.append(link)
        print("sukses")
    time.sleep(5)

driver.close()
```

```python
file = open("katadata_p2pLending.txt","w+")
print(len(links))
for link in links:
    print(link)
    file.write(link)
    file.write("\n")
```
