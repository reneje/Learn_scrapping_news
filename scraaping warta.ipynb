```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/Users/rinijannati/anaconda/bin/chromedriver", chrome_options=options)

driver = webdriver.Chrome()

url = "https://www.wartaekonomi.co.id/page/search/results/260064.html"
links=[]
driver.get(url)
hrefs = driver.find_elements_by_xpath("//*[@class='news-grid news-grid-row-3 clearfix padding-7px width-100']//*[@href]")
print('Number of results', len(hrefs))
for a in hrefs:
    link = a.get_attribute('href')
    print(link)
    links.append(link)
    print("sukses")

for i in range(2,6):
    url = "https://www.wartaekonomi.co.id/page/search/results/260064.html/"+str(i)
    driver.get(url)
    hrefs = driver.find_elements_by_xpath("//*[@class='news-grid news-grid-row-3 clearfix padding-7px width-100']//*[@href]")
    print('Number of results', len(hrefs))
    for a in hrefs:
        link = a.get_attribute('href')
        print(link)
        links.append(link)
        print("sukses")
    time.sleep(5)

driver.close()
```

```python
file = open("wartaekonomi_p2pLending.txt","w+")
print(len(links))
for link in links:
    print(link)
    file.write(link)
    file.write("\n")
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<strong.*?</strong>",' ',expr)
    expr = re.sub("<a>.*?</a>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
#     hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("[\n\t\r]","",hsl)
    return hsl
```

```python
from newspaper import Article

dates=[]
titles=[]
contents=[]
for link in links:
    if link is not None:
# link="http://www.tribunnews.com/bisnis/2019/02/22/ojk-usul-p2p-lending-pakai-layanan-asuransi-ini-alasannya"
        article = Article(link)
        article.download()
        page=requests.get(link)
        soup = BeautifulSoup(page.text,'html.parser')
# article.html
        cont = soup.find(class_='news-post-content clearfix')
        conti = cont.find_all('p')
        cont=bersihisi(cont)
        article.parse()
        dates.append(article.publish_date)
        titles.append(article.title)
        contents.append(cont)
        print(link)
        # print("dates: ", article.publish_date)
        # print("title: ", article.title)
        # print("text: ", article.text)
        print("sukses")
```
