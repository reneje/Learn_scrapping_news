```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import timez
from newspaper import Article
```

```python
links = open("okezone_link.txt").read().split("\n")
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<!-- paralax.*? paralax ads end -->",' ',expr)
    expr = re.sub("<div class=\"detail_tag\">.*?</div>"," ",expr)
    expr = re.sub("<script>.*?</script>"," ",expr)
    expr = re.sub("<p>Baca Selengkapnya.*?</strong>"," ",expr)
    expr = re.sub("<div style=\"font-family:.*?</a>"," ",expr)
    expr = re.sub("<!-- tag -->.*?</div>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\t"," ",hsl)
    hsl = re.sub('\(kmj\)','',hsl)
    hsl = re.sub("googletag.cmd.push.*?\}\);",'',hsl)
    return hsl
```

```python
titles=[]
dates=[]
contents=[]

for link in links:
    print(link)
    if link is not None:
        article = Article(link)
        article.download()
        article.parse()
        dates.append(article.publish_date)
        titles.append(article.title)
        contents.append(article.text)

        print("sukses")
```
