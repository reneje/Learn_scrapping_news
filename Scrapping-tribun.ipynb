```python
def bersihsub(expr):

    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("\t","",hsl)
    return hsl
```

```python
links=[]
hrefs = driver.find_elements_by_xpath("//*[@class='gs-title']//*[@href]")
print('Number of results', len(hrefs))

for a in hrefs:
    link = a.get_attribute('href')
    print(link)
    links.append(link)
```
