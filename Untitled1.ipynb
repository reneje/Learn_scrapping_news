```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<p>.*?</p>",' ',expr)
    expr = re.sub("<strong.*?</strong>",' ',expr)
    expr = re.sub("<a>.*?</a>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
#     hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("[\n\t\r]","",hsl)
    return hsl
```
