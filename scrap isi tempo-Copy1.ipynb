```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/Users/rinijannati/anaconda/bin/chromedriver", chrome_options=options)
```

```python
def bersihsub(expr):

    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("\t","",hsl)
    return hsl
```

```python
titles=[]
dates=[]
contents=[]
links_cont=[]
for link in links:
    if re.search("koran",link) or re.search("majalah",link):
        print(link)
        print("can't get content")
    else:
        print(link)
        page = requests.get(link)
        soup = BeautifulSoup(page.text,'html.parser')
        judul = soup.find(itemprop='headline')
        date = soup.find(itemprop="datePublished")
        title = bersihsub(judul)
        date = bersihsub(date)
        plist = soup.find(id="isi")
        isi = plist.findAll('p')
        isi = bersihsub(isi)
        links_cont.append(link)
        titles.append(titles)
        dates.append(date)
        contents.append(isi)
        print("sukses")
```

```python
file = open("tempo_p2pLending.txt","w+")
print(len(links))
for link in links:
    print(link)
    file.write(link)
    file.write("\n")
```
