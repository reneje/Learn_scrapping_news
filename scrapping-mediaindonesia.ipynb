```python
fh = open("mediaindonesia_link.txt","w")
link2=[]
links=list(set(links))
for link in links:
    if re.search("read/detail",link):
        print(link)
        link2.append(link)
        fh.write(link+"\n")
fh.close()
print(len(link2))
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub('<strong">.*?</strong>',' ',expr)
    expr = re.sub('<span class="tag".*?</span>',' ',expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
#     hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("[\n\t\r]","",hsl)
    return hsl
```

```python
# link ='https://www.merdeka.com/uang/satgas-waspada-investasi-temukan-182-entitas-p2p-lending-tak-berizin.html'
titles=[]
dates=[]
contents=[]
# links_isi=[]
for link in link2:
    print(link)
    try:
# link="https://mediaindonesia.com/read/detail/238342-p2p-lending-rambah-pembiayaan-pertanian-jagung"
# article = Article(link)
# article.download()
        page=requests.get(link)
        soup = BeautifulSoup(page.text,'html.parser')
        #         soup2 = soup.find(class_='col-custom left')
                # article.html
        sop = soup.find(class_="article-title")
        title = sop.find('h1')
        title = bersihisi(title)
        date = sop.find(class_='meta')
        #         date = date.find('span')
        date = bersihisi(date)
        cont = soup.find(itemprop='articleBody')
        cont = cont.findAll('p')
        cont = bersihisi(cont)

    #         article.parse()
        dates.append(date)
        titles.append(title)
        contents.append(cont)
#         print(link)
# print("dates: ", date)
# print("title: ", title)
# print("text: ", cont)
        print("sukses")
    except:
        dates.append("gagal")
        titles.append("gagal")
        contents.append("gagal")
        print("gagal")
```

```python
import pandas as pd
Link = pd.Series(link2)
Title = pd.Series(titles)
Date = pd.Series(dates)
Content = pd.Series(contents)

data = pd.DataFrame({"links":Link,"titles":Title,"dates":Date,"contents":Content})
# data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```
