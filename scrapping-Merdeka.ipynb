```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<div>*?</div>",' ',expr)
#     expr = re.sub("<ul>.*?</ul>",' ',expr)
#     expr = re.sub("<div class=\"detail_tag\">.*?</div>"," ",expr)
#     expr = re.sub("<script>.*?</script>"," ",expr)
#     expr = re.sub("<p>Baca Selengkapnya.*?</strong>"," ",expr)
#     expr = re.sub("<div style=\"font-family:.*?</a>"," ",expr)
#     expr = re.sub("<!-- tag -->.*?</div>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\t"," ",hsl)
#     hsl = re.sub('\(kmj\)','',hsl)
#     hsl = re.sub("googletag.cmd.push.*?\}\);",'',hsl)
    return hsl
```

```python
# link ='https://www.merdeka.com/uang/satgas-waspada-investasi-temukan-182-entitas-p2p-lending-tak-berizin.html'
titles=[]
dates=[]
contents=[]
links_isi=[]
for link in links:
    print(link)
    if re.search('foto',link):
        print("not get content")
    else:
        
        page=requests.get(link)
        soup = BeautifulSoup(page.text,'html.parser')
        judul = soup.find('h1')
        date = soup.find(class_='date-post')
        # date = rep.findAll('time')
        konten = soup.find(class_='mdk-body-paragpraph')
#         isi = konten.findAll('p')
        title = bersihisi(judul)
        date = bersihisi(date)
        isi = bersihisi(konten)
        titles.append(title)
        dates.append(date)
        contents.append(isi)
        links_isi.append(link)
        print("sukses")
#     print(title)
#     print(date)
#     print(isi)
```

```python
fh = open("merdeka_data_link.txt","w")
for link in links_isi:
    fh.write(link+"\n")
fh.close()
```

```python
import pandas as pd
raw_data = {"links":links_isi,"titles":titles,"dates":dates,"contents":contents}
data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```
