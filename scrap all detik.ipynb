```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<table.*?</table>",' ',expr)
    expr = re.sub("<div class=\"detail_tag\">.*?</div>"," ",expr)
    expr = re.sub("\ngoogletag.*?</script>"," ",expr)
    expr = re.sub("<a>.*?</a>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
#     hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("[\n\t]","",hsl)
    hsl = re.sub("googletag.cmd.push.*?\}\);",'',hsl)
    return hsl
```

```python
links_all=[]
titles=[]
dates=[]
contents=[]
for i in range (1,9):
    page2=requests.get("https://www.detik.com/search/searchall?query=p2p%20lending&siteid=2&sortby=time&page="+str(i))
    
    soup2 = BeautifulSoup(page2.text,'html.parser')
    linklist = soup2.find(class_='list media_rows list-berita')
    links = linklist.findAll('a')
    
    for i in range (0, len(links)):
        link = str(links[i]).split("\n")
        link_get = re.sub('<a href="','',link[0])
        link_get = re.sub('">','',link_get)
        print(link_get)
        if re.search('foto',link_get):
            print(link_get)
        else:
            links_all.append(link_get)
            page = requests.get(link_get)
            soup = BeautifulSoup(page.text,'html.parser')
            judul = soup.find('h1')
            date = soup.find(class_="date")
            cont = soup.find(class_='itp_bodycontent detail_text')
            cont = bersihisi(cont)
            title = bersihisi(judul)
            date = bersihisi(date)
            titles.append(title)
            dates.append(date)
            contents.append(cont)
            print('sukses')
```

```python
import pandas as pd
raw_data = {"links":links_all,"titles":titles,"dates":dates,"contents":contents}
data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```

```python
data.to_csv('detik_p2plending.csv',index=False)
```
