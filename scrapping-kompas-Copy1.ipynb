```python
def bersihsub(expr):

    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("  ","\n",hsl)
    return hsl
```

```python
# create empty array to store data

hrefs = driver.find_elements_by_xpath("//*[@class='gs-title']//*[@href]")
print('Number of results', len(hrefs))
links=[]
titles =[]
dates=[]
contents=[]
for a in hrefs:
    link = a.get_attribute('href')
    print(link)
    links.append(link)
    page = requests.get(link)
    soup = BeautifulSoup(page.text,'html.parser')
    judul = soup.find(class_ = 'read__title')
    time = soup.find(class_ = 'read__time')
    content = soup.find(class_ = 'read__content')
    judul = bersihsub(judul)
    time = bersihsub(time)
    content = bersihsub(content)
    date = re.sub("K.*? - ","",time)
    date = re.sub(",","",date)
#     links.append(link)
    titles.append(judul)
    dates.append(date)
    contents.append(content)
    print("Sukses")
    
print(len(links))
```
