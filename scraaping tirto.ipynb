```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/Users/rinijannati/anaconda/bin/chromedriver", chrome_options=options)

driver = webdriver.Chrome()

url = "https://tirto.id/search?q=p2p%20lending"
links=[]
driver.get(url)
hrefs = driver.find_elements_by_xpath("//*[@class='row my-5']//*[@href]")
print('Number of results', len(hrefs))
for a in hrefs:
    link = a.get_attribute('href')
    print(link)
    links.append(link)
    print("sukses")

for i in range(2,7):
    url = "https://tirto.id/search?q=p2p%20lending&p="+str(i)
    driver.get(url)
    hrefs = driver.find_elements_by_xpath("//*[@class='row my-5']//*[@href]")
    print('Number of results', len(hrefs))
    for a in hrefs:
        link = a.get_attribute('href')
        print(link)
        links.append(link)
        print("sukses")
    time.sleep(5)

driver.close()
```

```python
file = open("tirto_p2pLending.txt","w+")
print(len(links))
links=list(set(links))
for link in links:
    print(link)
    file.write(link)
    file.write("\n")
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub('<class="baca-holder bold".*?</div>',' ',expr)
    expt = re.sub('<div id="hide-cap">.*?</div',' ',expr)
#     expr = re.sub("<a>.*?</a>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
#     hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("[\n\t\r]","",hsl)
    return hsl
```

```python
from newspaper import Article

dates=[]
titles=[]
contents=[]
for link in links:
#     if link is not None:
# link="https://tirto.id/menggadai-data-diri-demi-ngutang-pinjaman-online-dgLB"
# article = Article(link)
# article.download()
    page=requests.get(link)
    soup = BeautifulSoup(page.text,'html.parser')
    # article.html
    title = soup.find(class_='news-detail-title text-center my-3 animated zoomInUp')
    title = bersihisi(title)
    date = soup.find(class_='detail-date mt-1 text-left')
    date = bersihisi(date)
    cont = soup.find(class_='col-12 content-detail-holder my-4')
    cont = bersihisi(cont)
# cont=bersihisi(cont)
#         article.parse()
    dates.append(date)
    titles.append(title)
    contents.append(cont)
    print(link)
#     print("dates: ", date)
#     print("title: ", title)
#     print("text: ", cont)
    print("sukses")
```

```python
import pandas as pd
Link = pd.Series(links)
Title = pd.Series(titles)
Date = pd.Series(dates)
Content = pd.Series(contents)

data = pd.DataFrame({"links":Link,"titles":Title,"dates":Date,"contents":Content})
# data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```
