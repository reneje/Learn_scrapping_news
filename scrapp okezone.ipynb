```python
def bersihsub(expr):

    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("\t","",hsl)
    return hsl
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<!-- paralax.*? paralax ads end -->",' ',expr)
    expr = re.sub("<div class=\"detail_tag\">.*?</div>"," ",expr)
    expr = re.sub("<script>.*?</script>"," ",expr)
    expr = re.sub("<p>Baca Selengkapnya.*?</strong>"," ",expr)
    expr = re.sub("<div style=\"font-family:.*?</a>"," ",expr)
    expr = re.sub("<!-- tag -->.*?</div>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\t"," ",hsl)
    hsl = re.sub('\(kmj\)','',hsl)
    hsl = re.sub("googletag.cmd.push.*?\}\);",'',hsl)
    return hsl
```
