```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub("<div class=\"baca-inline-head.*?</div>",' ',expr)
    expr = re.sub("<ul>.*?</ul>",' ',expr)
    expr = re.sub("<div class=\"detail_tag\">.*?</div>"," ",expr)
    expr = re.sub("<script>.*?</script>"," ",expr)
    expr = re.sub("<p>Baca Selengkapnya.*?</strong>"," ",expr)
    expr = re.sub("<div style=\"font-family:.*?</a>"," ",expr)
    expr = re.sub("<!-- tag -->.*?</div>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\t"," ",hsl)
    hsl = re.sub('\(kmj\)','',hsl)
    hsl = re.sub("googletag.cmd.push.*?\}\);",'',hsl)
    return hsl
```

```python
def bersihsub(expr):
    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("\t","",hsl)
    return hsl
```

```python
titles=[]
dates=[]
contents=[]
for link in links:
    print(link)
    page=requests.get(link)
    soup = BeautifulSoup(page.text,'html.parser')
    judul = soup.find('h1')
    date = soup.find('time')
    # date = rep.findAll('time')
    konten = soup.find(id='content')
    # isi = konten.findAll('p')
    title = bersihsub(judul)
    date = bersihisi(date)
    isi = bersihisi(konten)
    titles.append(title)
    dates.append(date)
    contents.append(isi)
    print("sukses")
    # print(title)
    # print(date)
    # print(isi)
```

```python
import pandas as pd
raw_data = {"links":links,"titles":titles,"dates":dates,"contents":contents}
data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```

```python
data.to_csv("sindonews_p2plending.csv",index=False)
```
