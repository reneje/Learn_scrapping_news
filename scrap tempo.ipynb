```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
```

```python
file = open("tempo_p2pLending.txt").read().split()
links = file
print(len(links))
```

```python
def bersihsub(expr):

    cleanr = re.compile('<.*?>')
    expr = str(expr)
    hsl = re.sub(cleanr,"",expr)
    hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("\t","",hsl)
    return hsl
```

```python
titles=[]
dates=[]
contents=[]
links_cont=[]
for link in links:
    if re.search("koran",link) or re.search("majalah",link):
        print(link)
        print("can't get content")
    else:
        print(link)
        page = requests.get(link)
        soup = BeautifulSoup(page.text,'html.parser')
        judul = soup.find(itemprop='headline')
        date = soup.find(itemprop="datePublished")
        title = bersihsub(judul)
        date = bersihsub(date)
        plist = soup.find(id="isi")
        isi = plist.findAll('p')
        isi = bersihsub(isi)
        links_cont.append(link)
        titles.append(title)
        dates.append(date)
        contents.append(isi)
        print("sukses")
```

```python
import pandas as pd
raw_data = {"links":links_cont,"titles":titles,"dates":dates,"contents":contents}
data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```

```python
data['titles'] = data['titles'].apply(lambda x : re.sub('\r','',x))
data
```
