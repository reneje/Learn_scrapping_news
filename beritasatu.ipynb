```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import time

options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')
options.add_argument('--headless')
driver = webdriver.Chrome("/Users/rinijannati/anaconda/bin/chromedriver", chrome_options=options)

driver = webdriver.Chrome()

# url = "https://www.beritasatu.com/search/?key=p2p+lending&type=article&page=1"
links=[]
# driver.get(url)
# hrefs = driver.find_elements_by_xpath("//*[@class='row my-5']//*[@href]")
# print('Number of results', len(hrefs))
# for a in hrefs:
#     link = a.get_attribute('href')
#     print(link)
#     links.append(link)
#     print("sukses")

for i in range(1,6):
    url = "https://www.beritasatu.com/search/?key=p2p+lending&type=article&page="+str(i)
    driver.get(url)
    hrefs = driver.find_elements_by_xpath("//*[@class='col-md-12 col-sm-12 kanal-media-outer']//*[@href]")
    print('Number of results', len(hrefs))
    for a in hrefs:
        link = a.get_attribute('href')
        print(link)
        links.append(link)
        print("sukses")
    time.sleep(5)

driver.close()
```

```python
def bersihisi(expr):
    expr = str(expr)
    expr = re.sub("\n"," ",expr)
    expr = re.sub('<class="baca-holder bold".*?</div>',' ',expr)
    expt = re.sub('<div id="hide-cap">.*?</div',' ',expr)
#     expr = re.sub("<a>.*?</a>"," ",expr)
    cleanr = re.compile('<.*?>')
    hsl = re.sub(cleanr,"",expr)
#     hsl = re.sub("\n"," ",hsl)
    hsl = re.sub("[\n\t\r]","",hsl)
    return hsl
```
