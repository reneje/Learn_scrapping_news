```python
from bs4 import BeautifulSoup
import urllib3
import requests
import re
import urllib.request
from selenium import webdriver
import time
from newspaper import Article
```

```python
links = open("katadata_p2pLending.txt").read().split("\n")
```

```python
links=list(set(links))
```

```python
links
```

```python
new_links=[]
titles=[]
dates=[]
contents=[]

for link in links:
    if re.search("search",link):
        print(link)
    else:
        new_links.append(link)
        article = Article(link)
        time.sleep(2)
        article.download()
        article.parse()
        dates.append(article.publish_date)
        titles.append(article.title)
        contents.append(article.text)
        print(link)
        print("sukses")
       
```

```python
contents
```

```python
import pandas as pd
raw_data = {"links":new_links,"titles":titles,"dates":dates,"contents":contents}
data = pd.DataFrame(raw_data,columns=["link","titles","dates","contents"])
data
```

```python
file = open("katadata_p2pLending2.txt","w+")
print(len(links))
for link in new_links:
    print(link)
    file.write(link)
    file.write("\n")
```

```python
data.to_csv("katadata_p2plending.csv",index=False)
```
